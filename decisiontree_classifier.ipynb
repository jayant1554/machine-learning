## 2.Decision tree  pre pruning and Hyperparameter tunning

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter={
    'criterion':['gini','entropy','log_loss'],
    "splitter":['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt'],
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
decisiontree=DecisionTreeClassifier()
clf=GridSearchCV(decisiontree,param_grid=parameter,cv=5,scoring="accuracy")
```

### for first dataset

```python
clf.fit(X_train,Y_train)
```

```python
clf.best_params_
```

```python
Y_pred=clf.predict(X_test)
```

```python
scoreC=accuracy_score(Y_pred,Y_test)
```

```python
print(scoreC)
print(classification_report(Y_pred,Y_test))
```
